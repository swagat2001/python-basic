```python
number_of_people = []
type_of_job = bank_df['job'].unique()
for i in type_of_job:
    con = bank_df['job'] == i
    countt = len(bank_df[con])
    number_of_people.append(countt)
```

```python
job_df = pd.DataFrame(zip(type_of_job,number_of_people),
                      columns = ['Job_type','Number_ofOpeople'])
job_df
```

```python
countt = []
for col in cat:
    labels = bank_df[col].unique()
    for i in labels:
        con = bank_df[col]==i
        length = len(bank_df[con])
        countt.append(length)
        df = pd.DataFrame(zip(labels,countt),
                         columns = ['column','count'])
        df.to_csv(f'{col}_df.csv',index = False)
```
