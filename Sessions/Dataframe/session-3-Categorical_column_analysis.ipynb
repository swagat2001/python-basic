```python
no_of_application=[]
labels = visa_df['continent'].unique()
for i in labels:
    con = visa_df['continent'] == i
    countt = len(visa_df[con])
    no_of_application.append(countt)
```

```python
no_of_application
```

```python
continent_df = pd.DataFrame(zip(labels,no_of_application),
                 columns = ['Continent','No applicants'])
continent_df
```

```python
continent_df.to_csv('continent.csv',index=False)
```

```python
countt =[]
for col in cat[1:]:
    labels = visa_df[col].unique()
    print(labels)
    for i in labels:
        con = visa_df[col] == i
        summ =len(visa_df[con])
        countt.append(summ)
        print(countt)
    df = pd.DataFrame(zip(labels,countt),columns = ['column','count'])
    df.to_csv(f'{col}_df.csv',index=False)
    
```

**Bar Charts**

- bar charts under the matplotlib


- plt.bar


- bar chart needs class names and class frequency


- class names we can get from value counts


- class frequency we can get from value counts

```python
continent_vc = visa_df['continent'].value_counts()
continent_vc
```

```python
keys = continent_vc.keys()
keys
```

```python
values = continent_vc.values
values
```

```python
plt.figure(figsize=(10,5))
colors = ['Red','Green','Yellow','Orange','Blue','cyan']
plt.bar(keys,values,color = colors)
plt.xlabel('class names')
plt.ylabel('class frequency')
plt.title('continent_BarChart')
plt.savefig('continent_BarChart.jpg')
plt.show()
```
