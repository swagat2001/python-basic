# **Data frame operations**

**Step-1: import packages**

- import pandas as pd

- import numpy as np

- import matplotlib.pyplot as plt

- import seaborn as sns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**step-2: create a dataframe using list**

```python
names = ['Ramesh','Suresh','Mahesh']
pd.DataFrame(names)
```

```python
names = ['Ramesh','Suresh','Mahesh']
age = [20,21,22]
pd.DataFrame(zip(names,age))
```

```python
names = ['Ramesh','Suresh','Mahesh']
age = [20,21,22]
city = ['Hyderabad','Banglore','Odisha']
pd.DataFrame(zip(names,age,city))
```

**step-3: Insert the columns**

- column value excpet an index or array-like

```python
names = ['Ramesh','Suresh','Mahesh']
pd.DataFrame(names,columns=['Names'])
```

```python
names = ['Ramesh','Suresh','Mahesh']
age = [20,21,22]
city = ['Hyderabad','Banglore','Odisha']
cols = ['Names','Age','City']
pd.DataFrame(zip(names,age,city),columns=cols)
```

**Step-4: Modify the index**

```python
names = ['Ramesh','Suresh','Mahesh']
age = [20,21,22]
city = ['Hyderabad','Banglore','Odisha']
cols = ['Names','Age','City']
idx = ['A','B','C']
df1 = pd.DataFrame(zip(names,age,city),index = idx,columns=cols)
```

```python
df1
```

```python
type(df1)
```

- index length should be match with number of rows


- columns length should be natch with number of cols

**Step-5: Create a Dataframe using dictionary**

```python
# names = ['Ramesh','Suresh','Mahesh']
# age = [20,21,22]
# city = ['Hyderabad','Banglore','Odisha']

d = {'Names':['Ramesh','Suresh','Mahesh'],
    'Age':[20,21,22],
    'City': ['Hyd','mumbai','pune']}
pd.DataFrame(d)

# keys behave as column names
# no need of explitly mention
```
